```python
# check how many ids are the same and how many different in train and test
train_ids = train_df['ID'].unique()
test_ids = test_df['ID'].unique()
print(len(train_ids))
print(len(test_ids))
# check how many ids are in both train and test
print(len(set(train_ids) & set(test_ids)))
# check how many ids are in train but not in test
print(len(set(train_ids) - set(test_ids)))
# check how many ids are in test but not in train
print(len(set(test_ids) - set(train_ids)))
# check how many ids are in train and test
print(len(set(train_ids) & set(test_ids)))
```
